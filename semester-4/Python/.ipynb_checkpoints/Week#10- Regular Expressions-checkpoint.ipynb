## The Regular Expression Module

```python
#regular expressions are used to match patterns in strings
#re module is used for regular expressions so we need to import it
import re
```

Before you can use regular expressions in your program, you must import the library using <b>“import re”</b>.

The simplest use of the regular expression library is the <b>search()</b> function.

You can use re.search() to see if a string matches a regular expression, similar to using the find() method for strings

### Using re.search() Like find()

<b>Example</b>: Search for lines that contain 'From':

```python
# Option#1:
import re
hand = open('texts.txt')

for line in hand:
    line = line.rstrip()
    
    if re.search('From:', line):
        print(line)
```

This program does not use the real power of regular expressions, since we could have just as easily used line.find() to accomplish the same result.

```python
# Option#2:
hand = open('texts.txt')
for line in hand:
    line = line.rstrip()
    if line.find('From:') >= 0:
        print(line)
```

The power of the regular expressions comes when we add special characters to the search string that allow us to more precisely control which lines match the string. Adding these special characters to our regular expression allow us to do sophisticated matching and extraction while writing very little code.

### Using re.search() Like startswith()

Write a program to only match lines where “From:” was at the <b>beginning</b> of the line:

```python
# Option#1:

import re

hand = open('texts.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line) :
        print(line)
```

This is still a very simple example that we could have done equivalently with the startswith() method from the string library.

```python
# Option#2:
hand = open('texts.txt')
for line in hand:
    line = line.rstrip()
    if line.startswith('From:') :
        print(line)
```

There are a number of other special characters that let us build even more powerful regular expressions. The most commonly used special character is the period or full stop, which matches any character.

<b>Example:</b> In the following example, the regular expression F..m: would match any of the strings “From:”, “Fxxm:”, “F12m:”, or “F!@m:” since the period characters in the regular expression match any character.

The <b>dot</b> character matches any character.

```python
# Search for lines that start with 'F', followed by 2 characters, followed by 'm:'
import re
hand = open('texts.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^F..m:', line):
        print(line)
```

## Wild-Card Characters

This is particularly powerful when combined with the ability to indicate that a character can be repeated any number of times using the * or + characters in your regular expression. <br>
These special characters mean that instead of matching a single character in the search string, they match zero-or-more characters (in the case of the asterisk) or one-or-more of the characters (in the case of the plus sign).

### Fine-Tuning Your Match

Depending on how “clean” your data is and the purpose of your application, you may want to narrow your match down a bit.

```python
X-Sieve: CMU Sieve 2.3
X-DSPAM-Result: Innocent
X-Plane is behind schedule: two weeks
X-: Very short
```

Example:

```python
# Search for lines that start with From and have an at sign
import re
hand = open('texts.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:.+@', line):
        print(line)
```

### Matching and Extracting Data

re.search() returns a True/False depending on whether the string matches  the regular expression

You can use re.findall() to extract portions of a string that match your regular expression, similar to a combination of find() and slicing: var[4:7] 

If we actually want the matching strings to be extracted, we use re.findall()

For example, we want to pull the email addresses from each of the following lines:<br>
From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008<br>
Return-Path: <postmaster@collab.sakaiproject.org><br>
          for <source@collab.sakaiproject.org>;<br>
Received: (from apache@localhost)<br>
Author: stephen.marquard@uct.ac.za

```python
import re
x= """From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008
Return-Path: postmaster@collab.sakaiproject.org
for source@collab.sakaiproject.org;
Received: (from apache@localhost)
Author: stephen.marquard@uct.ac.za"""
y = re.findall('\S+@\S+',x) 
print(y)
```

Some of our email addresses have incorrect characters like “<” or “;” at the beginning or end. Let’s declare that we are only interested in the portion of the string that starts and ends with a letter or a number.

```python
# Write your code here:
import re 
text = open('texts.txt')
for line in text:
    x =re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z0-9]',line)
    if len(x)>0:
        print(x)
```

Translating this regular expression, we are looking for substrings that start with a single lowercase letter, uppercase letter, or number “[a-zA-Z0-9]”, followed by zero or more non-blank characters (\S*), followed by an at-sign, followed by zero or more non-blank characters (\S*), followed by an uppercase or lowercase letter.

s -> {start with lowercase/uppercase/number}
ri.yogesh96 -> {zero or more non-blank characters}
@ -> {at-sign}
gmail.com -> {zero or more non-blank characters}

```python
import re
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('\d+', x)
print(y)
```

When we use re.findall(), it returns a list of zero or more sub-strings that match the regular expression.
for example we want to extract numbers from any line.

```python
import re
x = 'My 2 favorite numbers are 19 and 42.\nMy age is 32.'
print(x)
y = re.findall('[0-9]+',x)
print(y)
```

You can refine the match for re.findall() and separately determine which portion of the match is to be extracted by using parentheses

<b>Example:</b> If we want to find numbers on lines that start with the string “X-” such as:
<br>X-DSPAM-Confidence: 0.8475
<br>X-DSPAM-Probability: 0.0000

Translating this, we are saying, we want lines that start with X-, followed by zero or more characters (.*), followed by a colon (:) and then a space. After the space we are looking for one or more characters that are either a digit (0-9) or a period [0-9.]+.

```python
# write your code here
import re
text = open('texts.txt')
for line in text:
    x = re.findall('^X-.*: [0-9.]+',line)
    if len(x)>0:
        print(x)
```

Using paranthesis find numbers on lines that start with the string “X-”:

```python
import re
text = open('texts.txt')
for line in text:
    x = re.findall(r'^X-.*: ([0-9.]+)',line)
    if len(x)>0:
        print(x)
```

Parentheses are not part of the match, but they tell where to start and stop what string to extract.

```python
import re
x= 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('^From (\S+@\S+)',x)
print(y)
```

The repeat characters (* and +) push outward in both directions (greedy) to match the largest possible string

<img src='images/week13/greedy_matching.png'/>

<i>Source: https://www.py4e.com</i>

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+:', x)
print(y)
```

Not all regular expression repeat codes are greedy!  If you add a ? character, the + and * chill out a bit...

<img src='images/week13/not_greedy.png'/>

<i>Source: https://www.py4e.com</i>

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+?:', x)
print(y)
```

## Escape character

We can indicate that we want to simply match a character by prefixing that character with a backslash.

<b>Example:</b> we can find money amounts with the following regular expression.

```python
import re
x = 'We just received $10.00 for cookies.'
y=re.findall('\$([0-9.]+)',x)
print(y)
```
